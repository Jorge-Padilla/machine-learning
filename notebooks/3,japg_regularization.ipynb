```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import SGDRegressor
```

# Algorithm

```python
X = np.array([1,2,3,4,5,6])
y = np.array([1.5,2.8,3.8,3.7,5.7,5.5])

X_train = X[:2]
y_train = y[:2]

X_test = X[2:]
y_test = y[2:]
```

```python
plt.scatter(X_train,y_train, c='b', label='train')
plt.scatter(X_test,y_test, c='r', label='test')
plt.legend()
```

```python
vanilla = SGDRegressor(penalty=None)
ridge = SGDRegressor(penalty="l2", alpha=0.5)
lasso = SGDRegressor(penalty="l1", alpha=0.2)
```

```python
X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1)
X_test = X_test.reshape(-1,1)
y_test = y_test.reshape(-1)

vanilla.fit(X_train, y_train)
ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)
```

```python
y_hat_vanilla = vanilla.predict(X_test)
y_hat_ridge = ridge.predict(X_test)
y_hat_lasso = lasso.predict(X_test)

print(y_hat_vanilla)
print(y_hat_ridge)
print(y_hat_lasso)
```

```python
print(vanilla.score(X_test, y_test))
print(ridge.score(X_test, y_test))
print(lasso.score(X_test, y_test))
```

```python
X = X.reshape(-1,1)
plt.scatter(X, y, c='r')
plt.plot(X, vanilla.predict(X),label='vanilla')
plt.plot(X, ridge.predict(X),label='ridge')
plt.plot(X, lasso.predict(X),label='lasso')
plt.legend()
```
